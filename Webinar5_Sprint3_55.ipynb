```python
# Importamos
import pandas as pd
```

```python
# Cargamos el archivo

df_ts = pd.read_csv('taylor_swift_songs.csv')
```

```python
# Revisemoslo!

display(df_ts.head())
```

```python
# Un vistazo mas especifico
df_ts.info()
```

```python
# Ultimas 5 filas

display(df_ts.tail())
```

```python
# Seleccionar la fila con índice 5 (sexta fila)
#lista = [1,2,3,4,5,6,7]

#lista[4]

df_ts.loc[5]
```

```python
# Seleccionar las filas con índices 2, 5 y 10

df_ts.loc[[2,5,10]]
```

```python
# Seleccionar las filas con índices 2 al 10

df_ts.loc[2:10]
```

```python
# Seleccionar solo la columna 'Title'. Aquí tambien estoy generando el contenido de mi columna 'Title' para saber los nombre de los titulos de mis primero 5 registros.
df_ts.loc[0:4, "Title"]
```

```python
# Seleccionar las filas 0 a 4 y las columnas 'Title' y 'Album'

df_ts.loc[0:4], ["Title", "Album"]
```

```python
# Quiero saber todos los albumes que ha lanzado Taylor Swift
df_ts.loc[:, "Album"]
```

```python
# Filtrar canciones que pertenecen al álbum '1989'. Esto es una condicion que nos dice que debemos estructurar un dataframe que nos filter solo las canciones que pertenezcan al album 1989
df_ts.loc[df_ts["Album"] == "1989"]
```

```python
# Quiero saber que albumes fueron lanzados en el año 2020 y tuvieron más de 600 millones de reproducciones en Spotify
# Hint: Hacer uso de los condicionales.

df_ts.loc[(df_ts["Release Year"] == 2020) & (df_ts["Spotify Streams (millions)"] > 600)]
```

```python
# Filtrar canciones con reproducciones entre 500 y 1000 millones

df_ts.loc[(df_ts["Spotify Streams (millions)"] >= 500) & (df_ts["Spotify Streams (millions)"] <= 1000)]
```

```python
# Contar cuántos valores no nulos hay en cada columna
df_ts.isnull().sum()
```

```python
# Duración total de todas las canciones en minutos
df_ts["Duration (mins)"].sum()
```

```python
# Promedio de duración de canciones lanzadas después de 2015
df_ts.loc[df_ts["Release Year"] > 2015, "Duration (mins)"].mean()
```

```python
# Promedio de 'Duration (mins)' y 'Spotify Streams (millions)'
df_ts.loc[:, ["Duration (mins)", "Spotify Streams (millions)"]].mean()
```

```python
# Quiero saber las estadisticas de mis datos.

# Los quartiles nos permiten entender la distribución de nuestros datos.

df_ts.describe()
```
