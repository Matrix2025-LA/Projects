## Preparación de datos

```python
#Importar la librería pandas
import pandas as pd
```

```python
#Cargar el dataset
bank = pd.read_csv("/datasets/Churn.csv")
```

```python
#Visualizar las primeras filas y la información general del DataFrame
print(bank.head())
print(bank.info())
```

```python
#Eliminación de columnas que no aportan información relevante para el análisis
bank_clean = bank.drop(bank.columns[[0, 1, 2]], axis=1)
print(bank_clean.head())
```

```python
# Limpieza y transformación del DataFrame:
# - Normalización de los nombres de las columnas
# - Conversión de variables categóricas a formato numérico
# - Tratamiento de valores faltantes
bank_clean.columns = bank_clean.columns.str.lower()
bank_clean['geography'] = bank_clean['geography'].astype(str)
bank_clean = pd.get_dummies(bank_clean, columns=['geography'], drop_first=True)
bank_clean['gender'] = bank_clean['gender'].map({'Female': 0, 'Male': 1})
bank_clean['tenure'] = bank_clean['tenure'].fillna(bank_clean['tenure'].median())
print(bank_clean.head())
```

Conclusiones de la primera parte.


Se realizó una exploración inicial del DataFrame para conocer la estructura, los tipos de datos y la calidad de la información disponible. Durante este proceso, se identificaron las columnas RowNumber, CustomerId y Surname, las cuales fueron eliminadas debido a que no aportan valor significativo para el análisis ni para la construcción de modelos predictivos.

Asimismo, se estandarizaron los nombres de las columnas convirtiéndolos a minúsculas para facilitar su manipulación. Las columnas categóricas gender y geography fueron transformadas en variables numéricas mediante técnicas de codificación, permitiendo su uso en modelos de aprendizaje automático. Además, se trataron los valores faltantes en la columna tenure utilizando la mediana, con el fin de mantener la coherencia de los datos.

Como resultado de estos procesos, se obtuvo un nuevo DataFrame limpio, estructurado y listo para su posterior análisis y modelado.

## Pruebas iniciales del modelo sin corrección del desequilibrio de clases

```python
#Entrenamiento del modelo sin considerar el desequilibrio de clases
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.metrics import roc_curve, roc_auc_score

#División del dataset en conjuntos de entrenamiento (60%), validación (20%) y prueba (20%)
df_train, df_temp = train_test_split(bank_clean, test_size=0.40, random_state=12345)
df_valid, df_test = train_test_split(df_temp, test_size=0.50, random_state=12345)

#Separación de características (features) y variable objetivo (target)
features_train = df_train.drop(["exited"], axis=1)
target_train = df_train["exited"]

features_valid = df_valid.drop(["exited"], axis=1)
target_valid = df_valid["exited"]

features_test = df_test.drop(["exited"], axis=1)
target_test = df_test["exited"]

#Creación y entrenamiento del modelo de regresión logística
model = LogisticRegression(random_state=12345, solver="liblinear")
model.fit(features_train, target_train)

#Evaluación del modelo mediante accuracy en los conjuntos de entrenamiento y validación
score_train = model.score(features_train, target_train)
score_valid = model.score(features_valid, target_valid)

print("Accuracy del modelo de regresion en el conjunto de entrenamiento:", score_train)
print("Accuracy del modelo de regresion en el conjunto de validacion:", score_valid)

#Predicción sobre el conjunto de prueba y cálculo de la métrica F1-score
predicted_valid = model.predict(features_test)
print("F1", f1_score(target_test, predicted_valid))

#Cálculo de probabilidades para el valor AUC-ROC
predicted_proba = model.predict_proba(features_test)[:, 1]
auc_score = roc_auc_score(target_test, predicted_proba)
print("AUC-ROC:", auc_score)
```

Conclusiones.

El modelo entrenado sin aplicar técnicas de corrección para el desequilibrio de clases obtuvo una exactitud cercana al 78%, lo que podría interpretarse inicialmente como un buen desempeño. Sin embargo, al analizar métricas más representativas para problemas de clasificación desbalanceada, se observó un F1-score muy bajo (0.13) y un valor de AUC-ROC de 0.64.

Estos resultados indican que el modelo presenta una capacidad limitada para identificar correctamente la clase minoritaria, lo cual es crítico en este tipo de problemas. Por lo tanto, se concluye que es indispensable aplicar técnicas de balanceo de clases con el fin de mejorar la capacidad predictiva del modelo y obtener métricas más equilibradas y representativas.

## Corrección del modelo mediante técnicas de balanceo y evaluación del desempeño

## Modelo Random Forest aplicando la técnica de class_weight

```python
#Librerias a usar
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt

#Separación de características y variable objetivo
features = bank_clean.drop(["exited"], axis=1)
target = bank_clean["exited"]

#División del dataset en conjuntos de entrenamiento (75%) y validación (25%)
features_train, features_valid, target_train, target_valid = train_test_split(
    features, target, test_size=0.25, random_state=12345
)

#Creación del modelo Random Forest con balanceo automático de clases
model = RandomForestClassifier(n_estimators=100, random_state=12345, max_depth=12, class_weight='balanced')
model.fit(features_train, target_train)

#Predicciones y evaluación mediante F1-score
predicted_valid = model.predict(features_valid)
print("F1", f1_score(target_valid, predicted_valid))

#Cálculo de probabilidades para la curva ROC y la métrica AUC-ROC
predicted_proba = model.predict_proba(features_valid)[:, 1]
fpr, tpr, thresholds = roc_curve(target_valid, predicted_proba)
auc_score = roc_auc_score(target_valid, predicted_proba)
print("AUC-ROC:", auc_score)
```

```python
#Visualización de la curva ROC
plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, label=f'Random Forest (AUC = {auc_score:.3f})', linewidth=2)
plt.plot([0,1], [0,1], linestyle="--", color='gray', label='Clasificador aleatorio')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.xlabel("Tasa de falsos positivos")
plt.ylabel("Tasa de verdaderos positivos")
plt.title("Curva ROC - Random Forest")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

Conclusiones

Tras aplicar técnicas de balanceo mediante el parámetro class_weight='balanced' en el modelo Random Forest, se obtuvo una mejora significativa en el desempeño del modelo, especialmente en la capacidad para identificar la clase minoritaria. El incremento del F1-score evidencia un mejor equilibrio entre precisión y recall, lo cual resulta fundamental en problemas con datos desbalanceados.

Asimismo, el valor de AUC-ROC muestra una mejora notable con respecto al modelo sin balanceo, indicando una mayor capacidad discriminativa entre las clases.

## Modelo Decision Tree aplicando la técnica de downsampling

```python
#Revisión de la variable objetivo para identificar la clase mayoritaria
bank_clean['exited'].value_counts()
```

```python
#Librerias a usar
from sklearn.utils import shuffle
from sklearn.tree import DecisionTreeClassifier

#Separación de características y variable objetivo
features = bank_clean.drop(["exited"], axis=1)
target = bank_clean["exited"]

#División del dataset en conjuntos de entrenamiento (75%) y validación (25%)
features_train, features_valid, target_train, target_valid = train_test_split(
    features, target, test_size=0.25, random_state=12345
)

#Definición de la función downsample para reducir el tamaño de la clase mayoritaria
def downsample (features, target, fraction):
  features_zeros = features[target == 0]
  features_ones = features[target == 1]
  target_zeros = target[target == 0]
  target_ones = target[target == 1]

  #Reducción aleatoria de la clase mayoritaria
  features_zeros_downsampled = features_zeros.sample(frac=fraction, random_state=12345)
  target_zeros_downsampled = target_zeros.sample(frac=fraction, random_state=12345)

  #Unión de las clases balanceadas
  features_downsampled = pd.concat([features_zeros_downsampled, features_ones])
  target_downsampled = pd.concat([target_zeros_downsampled, target_ones])

  #Mezcla aleatoria de los datos
  features_downsampled, target_downsampled = shuffle(
      features_downsampled, target_downsampled, random_state=12345
  )
  return features_downsampled, target_downsampled

#Aplicación del downsampling al conjunto de entrenamiento
features_downsampled, target_downsampled = downsample(features_train, target_train, 0.5)

#Creación y entrenamiento del modelo Decision Tree
model = DecisionTreeClassifier(random_state=12345, min_samples_split=20, max_depth=50)
model.fit(features_downsampled, target_downsampled)

#Predicción y evaluación mediante F1-score
predicted_valid = model.predict(features_valid)
print("F1", f1_score(target_valid, predicted_valid))

#Cálculo del AUC-ROC para evaluar la capacidad discriminativa del modelo
predicted_proba = model.predict_proba(features_valid)[:, 1]
fpr, tpr, thresholds = roc_curve(target_valid, predicted_proba)
auc_score = roc_auc_score(target_valid, predicted_proba)
print("AUC-ROC:", auc_score)
```

```python
#Visualización de la curva ROC
plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, label=f'Decision Tree (AUC = {auc_score:.3f})', linewidth=2)
plt.plot([0,1], [0,1], linestyle="--", color='gray', label='Clasificador aleatorio')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.xlabel("Tasa de falsos positivos")
plt.ylabel("Tasa de verdaderos positivos")
plt.title("Curva ROC - Decision Tree")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

Conclusiones:

Se aplicó la técnica de downsampling con el objetivo de reducir el desequilibrio entre clases, disminuyendo el tamaño de la clase mayoritaria en el conjunto de entrenamiento. Esta estrategia permitió entrenar un modelo Decision Tree más equilibrado, favoreciendo una mejor identificación de la clase minoritaria.

Los resultados obtenidos muestran una mejora en el F1-score en comparación con el modelo entrenado sin técnicas de balanceo, lo que indica un mejor compromiso entre precisión y recall. Asimismo, el valor de AUC-ROC evidencia una capacidad aceptable del modelo para diferenciar entre ambas clases. No obstante, la reducción del tamaño del conjunto de entrenamiento puede ocasionar una pérdida de información, por lo que resulta conveniente comparar este enfoque con otras técnicas de balanceo, como el upsampling o el uso de pesos de clase.

## Evaluación final del modelo seleccionado: Random Forest con balanceo mediante class_weight

```python
#División del dataset en conjuntos de entrenamiento (60%), validación (20%) y prueba (20%)
df_train, df_temp = train_test_split(bank_clean, test_size=0.40, random_state=12345)
df_valid, df_test = train_test_split(df_temp, test_size=0.50, random_state=12345)

#Separación de características (features), variable objetivo (target) y pruebas (test)
features_train = df_train.drop(["exited"], axis=1)
target_train = df_train["exited"]

features_valid = df_valid.drop(["exited"], axis=1)
target_valid = df_valid["exited"]

features_test = df_test.drop(["exited"], axis=1)
target_test = df_test["exited"]

#Creación y entrenamiento del modelo final
final_model = RandomForestClassifier(n_estimators=100, random_state=12345, max_depth=12, class_weight='balanced')
final_model.fit(features_train, target_train)

#Evaluación del modelo en el conjunto de prueba mediante accuracy
final_accuracy = final_model.score(features_test, target_test)
print(f"Accuracy en conjunto de prueba: {final_accuracy}")

#Predicciones y evaluación mediante F1-score
predicted_valid = final_model.predict(features_test)
print("F1", f1_score(target_test, predicted_valid))

#Cálculo de la métrica AUC-ROC para evaluar la capacidad discriminativa del modelo
predicted_proba = final_model.predict_proba(features_test)[:, 1]
fpr, tpr, thresholds = roc_curve(target_test, predicted_proba)
auc_score = roc_auc_score(target_test, predicted_proba)
print("AUC-ROC:", auc_score)
```
